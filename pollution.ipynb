```python
# to get the top 5 countries with highest GDP per capita in 2007

df_2007 = df[df['year'] == 2007]
top_5_gdp = df_2007[['country', 'gdpPercap']].sort_values('gdpPercap', ascending=False).head(5)
print(top_5_gdp)
```

```python
# to get the average life expectancy for each country over all years

avg_lifeexp_country = df.groupby('country')['lifeExp'].mean()
print(avg_lifeexp_country)
```

```python
# to get the average life expectancy for each continent over all years

avg_lifeexp_continent = df.groupby('continent')['lifeExp'].mean()
print(avg_lifeexp_continent)
```

```python
# to get the total population for each country in 2007

df_2007 = df[df['year'] == 2007]
pop_2007 = df_2007.groupby('country')['pop'].sum()
print(pop_2007)
```

```python
continent_yearly_expectancy = df.groupby(['continent', 'year'])['lifeExp'].mean()
continent_yearly_expectancy.unstack().plot()
plt.xlabel('Year')
plt.ylabel('Life Expectancy (years)')
plt.title('Yearly Life Expectancy by Continent')
plt.show()
```

```python
continent_yearly_gdp = df.groupby(['continent', 'year'])['gdpPercap'].mean()
continent_yearly_gdp.unstack().plot()
plt.xlabel('Year')
plt.ylabel('GDP per capita')
plt.title('Yearly GDP per capita by Continent')
plt.show()
```
